```python
lloyd = {
 "name": "Lloyd",
 "homework": [90.0,97.0,75.0,92.0],
 "quizzes": [88.0,40.0,94.0],
 "tests": [75.0,90.0]
}

alice = {
 "name": "Alice",
 "homework": [100.0, 92.0, 98.0, 100.0],
 "quizzes": [82.0, 83.0, 91.0],
 "tests": [89.0, 97.0]
}

tyler = {
 "name": "Tyler",
 "homework": [0.0, 87.0, 75.0, 22.0],
 "quizzes": [0.0, 75.0, 78.0],
 "tests": [100.0, 100.0]
}

name_of_students = [lloyd, alice, tyler]

for students in name_of_students: 
    print(f'{students["name"]}')
    print(f'{students["homework"]}')
    print(f'{students["quizzes"]}')
    print(f'{students["tests"]}')

def average(numb): 
    total   = sum(numb)
    average = float(total)
    return average / len(numb)

def get_average(students):
    homework    = average(students["homework"])
    quizzes     = average(students["quizzes"])
    tests       = average(students["tests"])
    
    studentavrg = (homework * 10/100) + (quizzes * 30/100) + (tests * 60/100)
    return studentavrg

def get_letter_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"
    
print(get_letter_grade(get_average(lloyd))) 

def get_class_average(students):
    result = []

    for students in students:
        student_avrg = get_average(students)
        result.append(student_avrg)
    return average(result)

print(get_class_average(name_of_students))
print(get_letter_grade(get_class_average(name_of_students)))
```

```python
days = ("Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday")
sales = []

for perday in days :
    sales2 = float(input(f"Enter sales for {perday}:"))
    while sales2 <= 0:
        print(f"Input was invalid. Re-enter the sales for {perday}: ")
    
    else:
        sales.append(sales2)



total= sum(sales)
minimum= min(sales)
maximum= max(sales)

print(f"total sales : ${total:,.2f}")
print(f"minimum sale amount : ${minimum:,.2f}")
print(f"maximum sale amount : ${maximum:,.2f}")
```
